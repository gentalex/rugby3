```python
def fn_df_club(ville):

    liste_stats = []
    liste_nom = []
    liste_poste = []
    
    club = ville.lower()

    url = f'https://www.ultimaterugby.com/app/public/index.php/{dico_club[club]}/squad'

    html = requests.get(url)

    soup = BeautifulSoup(html.text, 'html.parser')

    test = soup.find_all('div', {"class" : "col-md-2 col-sm-4 col-xs-6"})

    for player in range(len(test)):
    # for player in range(3): # test avec 3 pour plus de lisibilité
        name = str(test[player].find_all('h4')[0].get_text(separator='<br/>')).replace("<br/>"," ")
        # print(name)
        liste_nom.append(name)
        poste = test[player].find_all('b')[0].get_text()
        # print(poste)
        liste_poste.append(poste)
        try:
            stats = ([test[player].find_all('li')[ele].get_text() for ele in range(len(test[player].find_all('li')))])
            # print(stats)
            if len(stats) == 3:
                # print(len(stats))
                liste_stats.append(stats)
            else:
                liste_stats.append(np.nan)
            # print('---')
        # print(test[player].find_all('b'))
        except IndexError:
            pass
        
    df_rugby_fn = pd.DataFrame({
        
        "Nom":liste_nom,
        "Poste":liste_poste,
        "Stats":liste_stats
    })
    # print(len(liste_nom))
    # print(len(liste_poste))
    # print(len(liste_stats))
    
    # Faire les modifications du df avant de le return
    # df_rugby_fn['Age'] = df_rugby_fn['Stats'].apply(lambda x : x[0])
    df_rugby_fn = df_rugby_fn.dropna()
    
    return df_rugby_fn.dropna().sample(3)
```
