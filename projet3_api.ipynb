```python
# date = input('Insérez une date au format (AAAA-MM-JJ) : ')

equipe = 'Stade Rochelais'

# url_top14 = f'https://api.sofascore.com/api/v1/sport/rugby/scheduled-events/{date}'
url_top14 = f'https://api.sofascore.com/api/v1/sport/rugby/scheduled-events/2024-11-30'

# Effectuer une requête à l'API
r = requests.get(url_top14)
data = r.json()

# pprint(data['events'][0])

# int(len(data['events']))

for team in range(len(data['events'])):
#     pprint(data['events'][away_teams]['awayTeam']['name'])
    if data['events'][team]['awayTeam']['name'] == equipe:
        # pprint(data['events'][away_teams]['awayTeam']['name'])
        num_match = team
    elif data['events'][team]['homeTeam']['name'] == equipe:
        # pprint(data['events'][away_teams]['awayTeam']['name'])
        num_match = team

# Score équipe à domicile
print(f"Equipe à domicile : {data['events'][num_match]['homeTeam']['name']}")        
pprint(data['events'][num_match]['homeScore']['normaltime'])
# Score équipe extérieur
print(f"Equipe à l\'extérieur : {data['events'][num_match]['awayTeam']['name']}")  
pprint(data['events'][num_match]['awayScore']['normaltime'])
```
